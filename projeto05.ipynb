```python
# Instala o Whisper open-source e o Gradio
!pip install git+https://github.com/openai/whisper.git gradio -q

import gradio as gr
import whisper

# Carrega o modelo na GPU do Colab (escolha: tiny, base, small, medium, large)
modelo = whisper.load_model("medium")

def transcrever_local(audio_filepath):
    if audio_filepath is None: return "Sem áudio."

    # Faz a transcrição usando a GPU local
    resultado = modelo.transcribe(audio_filepath)
    return resultado["text"]

# Interface
app = gr.Interface(
    fn=transcrever_local,
    inputs=gr.Audio(type="filepath"),
    outputs="text",
    title="App de Transcrição (Gratuito/Local)"
)

app.launch(share=True)
```
